## 1. Load Dataset

```python
dataset_path = "/Users/noura/Downloads/archive/UNSW_NB15_training-set.csv"
df = pd.read_csv(dataset_path)
print("Dataset loaded successfully!")
df.head()
```

## 4. Missing Values

```python

df.isnull().sum()
```

## 5. Target Variable Analysis

```python
if 'label' in df.columns:
    print("Target variable distribution:")
    print(df['label'].value_counts())
    
    df['label'].value_counts().plot(kind='bar')
    plt.title("Label Distribution (0 = Normal, 1 = Attack)")
    plt.xlabel("Label")
    plt.ylabel("Count")
    plt.show()
else:
    print("Target column 'label' not found.")
```

## 6. Correlation Between Numerical Features

```python

numeric_df = df.select_dtypes(include=['int64','float64'])

corr = numeric_df.corr()

plt.figure()
plt.imshow(corr)
plt.title("Correlation Matrix")
plt.colorbar()
plt.show()
```


## Conclusion

This EDA step helped us understand:

- Dataset size and structure
- Data types
- Missing values
- Target variable distribution
- Relationships between numerical features

These insights help prepare the dataset for machine learning in the next steps.


## 7. Distribution of Numerical Features

```python

import matplotlib.pyplot as plt

numerical_cols = df.select_dtypes(include=['int64','float64']).columns[:5]

for col in numerical_cols:
    plt.figure()
    df[col].hist()
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

## 8. Categorical Feature Analysis

```python

categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols[:3]:
    plt.figure()
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f"Top Categories in {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

## 9. Outlier Detection (Boxplots)

```python

for col in numerical_cols:
    plt.figure()
    plt.boxplot(df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

## 10. Feature vs Target Relationship

```python

if 'label' in df.columns:
    for col in numerical_cols:
        plt.figure()
        df.boxplot(column=col, by='label')
        plt.title(f"{col} vs Label")
        plt.suptitle("")
        plt.show()
```

## 11. Improved Correlation Heatmap

```python

numeric_df = df.select_dtypes(include=['int64','float64'])
corr = numeric_df.corr()

plt.figure(figsize=(10,8))
plt.imshow(corr)
plt.title("Correlation Heatmap")
plt.colorbar()
plt.show()
```
